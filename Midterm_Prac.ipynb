# Multiple Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

#Varying y-values and x-values
july_sales = [500,800,700,900,750,600,400]
august_sales = [800,900,600,800,1000,700,650]
days_of_the_week = [1,2,3,4,5,6,7]

#Setting x-labels
week_ticks = ['Week 1','Week 2','Week 3','Week 4','Week 5','Week 6','Week 7']

#Plotting the two different lines
plt.plot(days_of_the_week,july_sales,marker='o',color='b')
plt.plot(days_of_the_week,august_sales,marker='v',color='r')

#Putting a title and labels
plt.title("July Sales vs. August Sales Per Week")
plt.ylabel("Number of Sales")
plt.legend(['July Sales','August Sales'],loc = 'lower center')
plt.xticks(days_of_the_week,week_ticks)
plt.show()
```

# Scatter plot

```python
# We create a line of best-fit by using numpy's unique(), poly1d() , and polyfit() functions

# refer the above link for links of above functions
```

```python
#Getting x and y values
temperatures = [68,63,74,72,79,78,71,71,69,66,80,58]
customers = [317,355,463,419,507,431,433,434,362,340,540,250]


#Creating the scatter plot
plt.scatter(temperatures, customers, s=100, facecolors ='none', edgecolors='black', alpha=0.5)

#Labelling the plot
plt.xlabel('Average Temperature at the Pool (Degrees Fahrenheit)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers vs. Average Temperature at the Pool')
plt.plot(np.unique(temperatures),np.poly1d(np.polyfit(temperatures,customers,1))(np.unique(temperatures)))

#Outputting the scatter plot
plt.show()
```

# Subplot


```python
#Creating data
#First plot:
average_temp_in_june = [70,70,74,80]
customers_in_june = [237,250,325,457]

#Second plot:
average_temp_in_july = [76,78,80,82]
customers_in_july = [480,489,498,500]

#Third plot
average_temp_in_august = [71,73,80,86]
customers_in_august = [410,500,600,615]

#Creates one plot with random x & y values
#Creating subplot: subplot(nrows, ncols, index)
plt.subplot(2,2,1)
#Plotting
plt.plot(average_temp_in_june,customers_in_june)
plt.xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Creates a second plot with random x & y values
plt.subplot(2,2,2)
plt.plot(average_temp_in_july,customers_in_july)
plt.xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Creates a third plot with random x & y values
plt.subplot(2,1,2)
plt.plot(average_temp_in_august,customers_in_august)
plt.xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
plt.ylabel('Amount of customers',fontsize=8)

#Making room for the x-labels (https://stackoverflow.com/questions/6774086/why-is-my-xlabel-cut-off-in-my-matplotlib-plot)
plt.tight_layout()

#Puts all of the plots in the same figure 
plt.show()
```

```python
# This example uses the add_subplot function and has randomly-ordered data
#Therefore, it includes scatter plots instead of line plots

#Import package to help style graphs
from matplotlib import style

#Setting the style of the subplot
style.use('fivethirtyeight')
fig = plt.figure()  #figure bject

#Creating data
#First plot:
average_temp_in_june = [70,74,80,70]
customers_in_june = [250,325,457,237]

#Second plot:
average_temp_in_july = [82,76,78,80]
customers_in_july = [500,480,489,498]

#Third plot
average_temp_in_august = [80,71,86,73]
customers_in_august = [600,410,615,500]

#Returns the instance of the Axes. 
#E.g., fig.add_subplot(221) divides the figure into two rows (the first 2) and two columns (the second 2) and selected the first of 4 possible locations.
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

#Creates one plot with random x & y values
ax1.set_ylabel('Amount of customers',fontsize=8)
ax1.set_xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
ax1.scatter(average_temp_in_june,customers_in_june)

#Creates a second plot with random x & y values
ax2.set_ylabel('Amount of customers',fontsize=8)
ax2.set_xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
ax2.scatter(average_temp_in_july,customers_in_july)

#Creates a third plot with random x & y values
ax3.set_ylabel('Amount of customers',fontsize=8)
ax3.set_xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
ax3.scatter(average_temp_in_august,customers_in_august)

#Labelling the subplots and creating space 
fig.suptitle('Average Temperatures vs. Amount of Customers at the Pool',fontsize=12)
fig.subplots_adjust(hspace=1.0,top=1.0,wspace=1.0)
fig.tight_layout()

#Puts all of the plots in the same figure 
plt.show()
```
